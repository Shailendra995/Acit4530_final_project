```python
df = pd.read_csv('Accuracy_each_activities.csv')
```

```python
X = df.iloc[:, :-1].values # We take all the feature columns (we exclude the last one)
#Y = df.iloc[:, 5].values # We take the dependent variables
df
```

```python
X = df['Generation']
A = df['Gate3']
B = df['Gate5']
C = df['Gate10']
D = df['Gate15']
E = df['Gate20']

F = df['std3']
G = df['std5']

#bin_vector = np.loadtext('sacc_or_fixation.txt')
plt.plot(X[0:1000],A[0:500], c='b', label= '3-Gates')
plt.plot(X[0:1000],B[0:500], c='g', label ='5-Gates')
plt.plot(X[0:1000],C[0:500], c='r', label ='10-Gates')
plt.plot(X[0:1000],D[0:500], c='c', label ='15-Gates')
plt.plot(X[0:1000],E[0:500], c='m', label ='20-Gates')
plt.legend(loc="upper right")
plt.title('')
plt.xlabel('No.of generation',fontsize=15)
plt.ylabel('Fitness Score',fontsize=15)
plt.savefig('DIFF_FitnessVsgen_gates_10P_Mut.pdf')
plt.show()
```

```python
#for standard deviation

Gate3 = df["Gate3"]
Gate5 = df['Gate5']
Gate10 = df['Gate10']
Gate15= df['Gate15']
Gate20 = df['Gate20']


std_1 = Gate3.std();
print(std_1);

std_2 =Gate5.std();
print(std_2);

std_3 = Gate10.std();
print(std_3);

std_4 = Gate15.std();
print(std_4);

std_5 = Gate20.std();
print(std_5);
```

```python

x = df["Generation"]
plt.plot(x, A, 'r-', label='3-Gates')
plt.fill_between(x, A-std_1 ,  A +std_1 , color='r', alpha=0.2)

plt.plot(x[0:2000], B[0:1000], 'g-', label='5-Gates',alpha=1)
plt.fill_between(x, B-std_2 ,  B+std_2 , color='g', alpha=0.2)

plt.plot(x, C, 'b-', label='10-Gates')
plt.fill_between(x, C-std_3 ,  C+std_3 , color='b', alpha=0.2)


plt.plot(x, D, 'm-', label='15-Gates',alpha=1)
plt.fill_between(x, D-std_4 ,  D+std_4 , color='m', alpha=0.1)

plt.plot(x, E, 'k-', label='20-Gates')
plt.fill_between(x, E-std_5 ,  E+std_5 , color='k', alpha=0.1)


plt.legend()

plt.xlabel('No.of generation',fontsize=15)
plt.ylabel('Fitness Score',fontsize=15)
plt.savefig('Diff_FitnessVsgenSTD_mut_Random_KLfitness_15gates.pdf')
plt.show()
```
